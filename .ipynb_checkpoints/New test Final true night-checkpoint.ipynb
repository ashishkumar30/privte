````python
class ICAFile:
    def __init__(self):
        self.week = 0
        self.ica_number = ""
        self.store = ""
        self.st_override = '0'
        self.event = 0
        self.system = ""
        self.vendor = ""
        self.item = ""
        self.generation = ''
        self.upc_man_override = ''
        self.keycat = 0
        self.dominance_code = ''
        self.source_code = ''
        self.description = ''
        self.qc_flag = ''
        self.ica_recipe = [0] * 25
        self.ica_var_length = 0
        self.ica_variables = ''

    def __str__(self):
        return (f"Week: {self.week}, ICA Number: {self.ica_number}, Store: {self.store}, "
                f"ST Override: {self.st_override}, Event: {self.event}, System: {self.system}, "
                f"Vendor: {self.vendor}, Item: {self.item}, Generation: {self.generation}, "
                f"UPC Man Override: {self.upc_man_override}, Keycat: {self.keycat}, "
                f"Dominance Code: {self.dominance_code}, Source Code: {self.source_code}, "
                f"Description: {self.description}, QC Flag: {self.qc_flag}, "
                f"ICA Recipe: {self.ica_recipe}, ICA Var Length: {self.ica_var_length}, "
                f"ICA Variables: {self.ica_variables}")


def parse_ica_record(line):
    ica_record = ICAFile()
    print(f"Parsing line: {line}")  # Debugging statement
    try:
        ica_record.week = int(line[0:4].strip())
    except ValueError:
        ica_record.week = 0

    ica_record.ica_number = line[4:10].strip()
    ica_record.store = line[10:15].strip()
    ica_record.st_override = line[15:16].strip()
    try:
        ica_record.event = int(line[16:20].strip())
    except ValueError:
        ica_record.event = 0

    ica_record.system = line[20:22].strip()
    ica_record.vendor = line[22:28].strip()
    ica_record.item = line[28:34].strip()
    ica_record.generation = line[34:35].strip()
    ica_record.upc_man_override = line[35:36].strip()
    try:
        ica_record.keycat = int(line[36:40].strip())
    except ValueError:
        ica_record.keycat = 0
    ica_record.dominance_code = line[40:41].strip()
    ica_record.source_code = line[41:42].strip()
    ica_record.description = line[42:74].strip()
    ica_record.qc_flag = line[74:75].strip()
    # Assuming rest of the line contains ica_recipe, ica_var_length, and ica_variables
    return ica_record


def process_microwand_data(input_file, output_file):
    with open(input_file, 'r') as infile:
        records = []
        for line in infile:
            if line.strip():  # Ignore empty lines
                record = parse_ica_record(line)
                records.append(record)

    with open(output_file, 'w') as outfile:
        for record in records:
            outfile.write(str(record) + '\n')


def main():
    input_file = 'ICA.W2373.U0221.S14409.D250218.T082530'
    output_file = 'ICA_UK_Converted_Data.txt'
    process_microwand_data(input_file, output_file)


if __name__ == "__main__":
    main()
````

````python
import pandas as pd

class ICAFile:
    def __init__(self):
        self.week = 0
        self.ica_number = ""
        self.store = ""
        self.st_override = '0'
        self.event = 0
        self.system = ""
        self.vendor = ""
        self.item = ""
        self.generation = ''
        self.upc_man_override = ''
        self.keycat = 0
        self.dominance_code = ''
        self.source_code = ''
        self.description = ''
        self.qc_flag = ''
        self.ica_recipe = [0] * 25
        self.ica_var_length = 0
        self.ica_variables = ''

    def __str__(self):
        return (f"Week: {self.week}, ICA Number: {self.ica_number}, Store: {self.store}, "
                f"ST Override: {self.st_override}, Event: {self.event}, System: {self.system}, "
                f"Vendor: {self.vendor}, Item: {self.item}, Generation: {self.generation}, "
                f"UPC Man Override: {self.upc_man_override}, Keycat: {self.keycat}, "
                f"Dominance Code: {self.dominance_code}, Source Code: {self.source_code}, "
                f"Description: {self.description}, QC Flag: {self.qc_flag}, "
                f"ICA Recipe: {self.ica_recipe}, ICA Var Length: {self.ica_var_length}, "
                f"ICA Variables: {self.ica_variables}")


def parse_ica_record(line):
    ica_record = ICAFile()
    print(f"Parsing line: {line}")  # Debugging statement
    try:
        ica_record.week = int(line[0:4].strip())
    except ValueError:
        ica_record.week = 0

    ica_record.ica_number = line[4:10].strip()
    ica_record.store = line[10:15].strip()
    ica_record.st_override = line[15:16].strip()
    try:
        ica_record.event = int(line[16:20].strip())
    except ValueError:
        ica_record.event = 0

    ica_record.system = line[20:22].strip()
    ica_record.vendor = line[22:28].strip()
    ica_record.item = line[28:34].strip()
    ica_record.generation = line[34:35].strip()
    ica_record.upc_man_override = line[35:36].strip()
    try:
        ica_record.keycat = int(line[36:40].strip())
    except ValueError:
        ica_record.keycat = 0
    ica_record.dominance_code = line[40:41].strip()
    ica_record.source_code = line[41:42].strip()
    ica_record.description = line[42:74].strip()
    ica_record.qc_flag = line[74:75].strip()
    # Assuming rest of the line contains ica_recipe, ica_var_length, and ica_variables
    return ica_record


def process_microwand_data(input_file, output_file):
    records = []
    with open(input_file, 'r') as infile:
        for line in infile:
            if line.strip():  # Ignore empty lines
                record = parse_ica_record(line)
                records.append(record)

    # Convert records to a DataFrame
    df = pd.DataFrame([record.__dict__ for record in records])
    
    # Save DataFrame to a CSV file
    df.to_csv(output_file, index=False)

    return df


def main():
    input_file = 'ICA.W2373.U0221.S14409.D250218.T082530'
    output_file = 'ICA_UK_Converted_Data.csv'
    df = process_microwand_data(input_file, output_file)
    print(df)


if __name__ == "__main__":
    main()
````
